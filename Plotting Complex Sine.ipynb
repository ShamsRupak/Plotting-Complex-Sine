# Plotting Complex Sine

## Write Python code (in a Jupyter notebook) that generates the same plots as the attached MATLAB script, using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define the range of x values
x = np.arange(-np.pi, np.pi, 0.05)
X, Y = np.meshgrid(x, x)
```

```python
# Compute Z values
Z = np.sin(X + 1j * Y)
```

```python
# Plot Real part of Z
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, np.real(Z), cmap='viridis')
ax.set_title('Real part')
ax.set_xlabel('x')
ax.set_ylabel('iy')
plt.show()
```

```python
# Plot Imaginary part of Z
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, np.imag(Z), cmap='viridis')
ax.set_title('Imaginary part')
ax.set_xlabel('x')
ax.set_ylabel('iy')
plt.show()
```

```python
# Plot Magnitude of Z
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, np.abs(Z), cmap='viridis')
ax.set_title('Magnitude')
ax.set_xlabel('x')
ax.set_ylabel('iy')
plt.show()
```
